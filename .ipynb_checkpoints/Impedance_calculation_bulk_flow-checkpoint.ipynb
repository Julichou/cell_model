```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

## Aim of the notebook

The aim of this notebook is to calculate different values of impedance. This impedance can be calculated in a bulk containing cells or in a microfluidic flow while cells passing one-by-one in front of the electrodes.  
Two models can be selected: a one shell-model, meaning that the cell is only represented by its cytosol and membrane, and a two-shell model, meaning the cell is represented by its cytosol, membrane, and wall.

## References

All the following formulas are derived from:

"MyDEP: A new computational tool for dielectric modeling of particles and cells" (Cottet et al, 2019)  
"High-speed single-cell dielectric spectroscopy" (Spencer and Morgan, 2020)  
"Recent advances in electrical impedance sensing technology for single-cell analysis" (Zhang et al, 2021)  
"A review of impedance measurements of whole cells" (Xu et al, 2016)

```python
## All the frequency that will be tested

f = np.arange(10e2, 10e7, 500)
```

```python
'''Function that calculates the reference impedance in absence of cell in bulk or flow'''

def ref_impedance_simulator(bulk_flow, vaccuum_permit, f, med_permit, med_elec_conductivity, S, d, w, h, l):
    imp = []
    
    if bulk_flow != 0 and bulk_flow != 1:
        print("Invalid value for bulk_flow, enter 0 or 1")
        return False
    if bulk_flow == 1:
        ## Cell constant
        k = w/h

    for i in f:
        ## Field angular frequency 
        ang_freq = 2*math.pi*i
    
        ## Calculating the particle and medium complex permittivity
        med_complx_permit = med_permit*vaccuum_permit-1j*(med_elec_conductivity/ang_freq)
 
        ## Calculating the complex capacitance of the system 
        if bulk_flow == 0:
            med_complx_cap = med_complx_permit*(S/d)
        if bulk_flow == 1:
            med_complx_cap = med_complx_permit*l*k 
    
        ## Calculating the complex impedance
        imp_complx = 1/(1j*ang_freq*med_complx_cap)

        imp.append(imp_complx)
    
    return imp
```

```python
'''function that calculates the normalized impedance (single- or two-shell model, in bulk or in flow)'''

### List of variables: ###

#bulk_flow: 0 if the impedance must be calculated in bulk, 1 if in flow
#shell_num: the number of shells of the model (1 or 2)
# cyto_permit: permittivity of cytosol  
# cyto_elec_conductivity: electrical conductivity of cytosol
# vaccuum_permit: vaccuum permittivity
# f: frequency
# med_permit: permittivity of medium
# med_elec_conductivity: electrical conductivity of medium
# mb_elec_conductivity: electrical conductivity of membrane
# mb_permit: permittivity of membrane
# mb_width: width of membrane in micrometers
# wall_width: wall width (for 2-shell model), put 0 if one-shell model
# wall_permit: wall permittivity (for 2-shell model), put 0 if one shell model
# wall_elec_conductivity: wall electrical conductivity (for 2-shell model), put 0 if one-shell model
# part_radius: radius of the cell in micrometers
# S: electrode surface in .... (only for bulk, enter 0 if calculation in flow)
# cell_concentration: number of cells/mL
# d: distance between electrodes in micrometers (only for bulk, enter 0 if calculation in flow)
# w: electrode width in micrometers (only for flow, enter 0 if calculation in bulk)
# h: channel heigh in micrometers (only for flow, enter 0 if calculation in bulk)
# l: length of electrodes in micrometers (only for flow, enter 0 if calculation in bulk)

def impedance_simulator(bulk_flow,
                        shell_num,
                        cyto_permit,
                        cyto_elec_conductivity,
                        vaccuum_permit,
                        f,
                        med_permit,
                        med_elec_conductivity,
                        mb_elec_conductivity,
                        mb_permit,
                        mb_width,
                        wall_width,
                        wall_permit,
                        wall_elec_conductivity,
                        part_radius,
                        S,
                        cell_concentration,
                        d,
                        w,
                        h,
                        l):
    
    if bulk_flow != 0 and bulk_flow != 1:
        print("Invalid value for bulk_flow, enter 0 or 1")
        return False
    
    imp = [] # the list of the complex impedance according to the frequency  
    ref_imp = [] # the list of reference impedance (without cell)
    
    if bulk_flow == 0:
        ## Calculating vol_frac
        cell_volume = (4/3)*math.pi*(part_radius)**3 #volume of one cell in micrometer^3
        total_cell_volume = cell_volume*cell_concentration #total volume occupied by the cells par mL
        vol_frac = total_cell_volume/10e12 #ratio between the volume occupied by the cells in one mL over one mL

        if vol_frac > 0.1:
            print("The fraction volume ({}) is higher than 0.1, while it must be lower than 0.1 to apply this function. The calculations cannot be done.".format(vol_frac))
            return False
        
    elif bulk_flow == 1:
        ## Cell constant
        k = w/h
    
        ## Calculating vol_frac
        vol_frac = (4/3)*math.pi*(part_radius)**3*(1/(k*w*l*h))
        
        if vol_frac > 0.4:
            print("The fraction volume ({}) is higher than 0.4, while it must be lower than 0.1 to apply this function. The calculations cannot be done.".format(vol_frac))
            return False
    
    for i in f:
        ## Field angular frequency 
        ang_freq = 2*math.pi*i
    
        ## Calculating the cytoplasm, medium, and membrane complex permittivity
        cyto_complx_permit = cyto_permit*vaccuum_permit-1j*(cyto_elec_conductivity/ang_freq)
        med_complx_permit = med_permit*vaccuum_permit-1j*(med_elec_conductivity/ang_freq)
        mb_complx_permit = mb_permit*vaccuum_permit-1j*(mb_elec_conductivity/ang_freq)
        wall_complx_permit = wall_permit*vaccuum_permit-1j*(wall_elec_conductivity/ang_freq)
    
        ## Calculating the particle perimttivity based on the shell model selected
        if shell_num == 1:
            part_complx_permit = mb_complx_permit*(((part_radius/(part_radius-mb_width))**3 +2*((cyto_complx_permit-mb_complx_permit)/(cyto_complx_permit+2*mb_complx_permit)))/
                                                   ((part_radius/(part_radius-mb_width))**3 - ((cyto_complx_permit-mb_complx_permit)/(cyto_complx_permit+2*mb_complx_permit))))
        elif shell_num == 2:
            # first calculate the equivalent of a one-shell model with just the cytosol and the membrane, 
            #then calculate the second shell based on the first shell and the cell wall
            one_shell_complx_permit = mb_complx_permit*((((part_radius-wall_width)/(part_radius-wall_width-mb_width))**3+2*((cyto_complx_permit-mb_complx_permit)/(cyto_complx_permit+2*mb_complx_permit)))/
                                                        (((part_radius-wall_width)/(part_radius-wall_width-mb_width))**3-((cyto_complx_permit-mb_complx_permit)/(cyto_complx_permit+2*mb_complx_permit))))
            part_complx_permit = wall_complx_permit*(((part_radius/(part_radius-wall_width))**3+2*((one_shell_complx_permit-wall_complx_permit)/(one_shell_complx_permit+2*wall_complx_permit)))/
                                                     ((part_radius/(part_radius-wall_width))**3-((one_shell_complx_permit-wall_complx_permit)/(one_shell_complx_permit+2*wall_complx_permit))))
        else:
            print("The algorithm cannot calculate more than a two-shell model, please choose 1 or 2 for shell_num")
            return False
       
        ## Calculating the Clausius-Mossotti factor
        CM_factor = (part_complx_permit-med_complx_permit)/(part_complx_permit+2*med_complx_permit)
       
        ## Calculating the complex permittivity of mixture cell-medium
        mix_complx_permit = med_complx_permit*((1+2*vol_frac*CM_factor)/(1-vol_frac*CM_factor))
    
        ## Calculating the complex capacitance of the system 
        if bulk_flow == 0:
            mix_complx_cap = mix_complx_permit*(S/d) 
        elif bulk_flow == 1:
            mix_complx_cap = mix_complx_permit*l*k 
    
        ## Calculating the complex impedance
        imp_complx = 1/(1j*ang_freq*mix_complx_cap)

        imp.append(imp_complx)
    
    # Get the reference impedance (without cells)
    ref_imp = ref_impedance_simulator(bulk_flow, vaccuum_permit, f, med_permit, med_elec_conductivity, S, d, w, h, l)

    ## Normalize the impedance
    Normalized_imp = [((element1 - element2)/element2) for (element1, element2) in zip(imp, ref_imp)]
    
    return Normalized_imp
```

```python
''' Plotting the impedance as a function of the frequency '''

# bulk_flow: 0 if we want the impedance in bulk, 1 if in flow
# Imp: the variable containing the normalized impedance

def plot_imp(bulk_flow, Imp):
    if bulk_flow == 0:
        condition = "bulk"
    else:
        condition = "flow"
        
    values = [[element.real for element in Imp], 
               [element.imag for element in Imp], 
               [abs(element) for element in Imp]] # to separate the real, imaginary and modulus parts of Imp
    title = "Modulus of normalized {} impedance".format(condition) # to label one of the axes
    
    ## Plot the normalized impedance
    fig, ax = plt.subplots(1,2,figsize= (15,6), sharey = False, sharex = False)
    sbplt = ax[0] # Plot the real and imaginary parts of the impedance on the same plot
    sbplt.plot(f, values[0])
    sbplt.plot(f, values[1])
    sbplt.legend(["Real", "Imag"])
    sbplt.set_xscale("log")
    sbplt.set_ylabel("Real and Imaginary parts of {} impedance (V)".format(condition))
    sbplt = ax[1] # Plot the modulus on another plot
    sbplt.plot(f, values[2])
    sbplt.set_xscale("log")
    sbplt.set_ylabel("{} (V)".format(title))
    
    fig.tight_layout()

    fig.text(0.47,0, "Frequency (Hz)")
    
    #return fig
```

```python
''' The function that retuns directly the plot of the nomalized impedance '''

def IMPEDANCE_SIMULATOR(bulk_flow,shell_num, cyto_permit, cyto_elec_conductivity, vaccuum_permit, f, 
                        med_permit, med_elec_conductivity, mb_elec_conductivity, mb_permit, mb_width,
                        wall_width, wall_permit, wall_elec_conductivity,part_radius, S, cell_concentration,
                        d,w, h,l):
    Imp = impedance_simulator(bulk_flow,shell_num,cyto_permit, cyto_elec_conductivity, vaccuum_permit, f, 
                               med_permit, med_elec_conductivity, mb_elec_conductivity, mb_permit, mb_width,
                               wall_width, wall_permit, wall_elec_conductivity,part_radius, S, cell_concentration,
                               d,w, h,l)
    fig = plot_imp(bulk_flow, Imp)
    return fig
```

```python
## Plotting the normalized impedance for bulk

IMPEDANCE_SIMULATOR(bulk_flow=0,
                    shell_num=2,
                    cyto_permit=50, 
                    cyto_elec_conductivity=0.3, 
                    vaccuum_permit=8.9e-12, 
                    f=f, 
                    med_permit=78, 
                    med_elec_conductivity=0.546, 
                    mb_elec_conductivity=2.5e-7, 
                    mb_permit=6, 
                    mb_width=8e-9,
                    wall_width=2.2e-7, 
                    wall_permit=60, 
                    wall_elec_conductivity=2.4e-2,
                    part_radius=3.5e-6, 
                    S=9e-10, 
                    cell_concentration=10**20, 
                    d=2e-2,
                    w=0,
                    h=0,
                    l=0)
```
